```python
#Was ist verbessert?

#Kategorien statt Rohwerte → übersichtlicher.
#Durchschnittswerte → klare Aussage.
#Farbpalette & Beschriftungen → professioneller Look.
#Werte auf Balken → sofort verständlich.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Beispiel-Daten (in deiner Anwendung: df = pd.read_csv('deine_datei.csv'))
data = {
    'Obesity levels (Country)': [15, 22, 35, 28, 42, 18, 33, 45, 20, 38],
    'Outdoor activities (City)': [80, 70, 50, 65, 40, 75, 55, 35, 72, 48]
}
df = pd.DataFrame(data)

# 1. Obesity Levels in Kategorien gruppieren
bins = [0, 20, 30, 40, 100]
labels = ['Niedrig', 'Mittel', 'Hoch', 'Sehr hoch']
df['Obesity_Group'] = pd.cut(df['Obesity levels (Country)'], bins=bins, labels=labels)

# 2. Durchschnittliche Outdoor-Aktivitäten pro Gruppe berechnen
grouped = df.groupby('Obesity_Group')['Outdoor activities (City)'].mean().reset_index()

# 3. Visualisierung
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10,6))
sns.barplot(x='Obesity_Group', y='Outdoor activities (City)', data=grouped, palette='viridis')

# Titel und Achsenbeschriftungen
plt.title('Durchschnittliche Outdoor-Aktivitäten nach Adipositas-Kategorie', fontsize=16, fontweight='bold')
plt.xlabel('Adipositas-Kategorie', fontsize=14)
plt.ylabel('Ø Outdoor-Aktivitäten', fontsize=14)

# Werte auf den Balken anzeigen
for index, row in grouped.iterrows():
    plt.text(index, row['Outdoor activities (City)'] + 1, f"{row['Outdoor activities (City)']:.1f}",
             ha='center', fontsize=12, color='black')

plt.tight_layout()
plt.show()
```
